```python
##Data Preprocessing##
```

```python
# all non-null values are replaced by digit 1 in dependents column so it has 0 null values now

df['Dependents'].fillna(1, inplace = True)
df.info()
```

```python
#All null values in Loanamount section is filled by the mean value of its own column to obtain 614 non-null values 

df['LoanAmount'].fillna(df.LoanAmount.mean(), inplace = True)
df.info()
```

```python
df.head(10)
```

```python
# creating a duplicate column of a column with binary outcomes in character form and appending it in the table as shown

Value_Mapping1 = {'Yes' :1, 'No' : 0}
df['Married_Section'] = df['Married'].map(Value_Mapping1)
df.head(10)
```

```python
# creating a duplicate column of a column with binary outcomes in character form and appending it in the table as shown

Value_Mapping2 = {'Male' :1, 'Female' : 0}
df['Gender_Section'] = df['Gender'].map(Value_Mapping2)
df.head(10)
```

```python
df['Education'].unique()
```

```python
# creating a duplicate column of a column with binary outcomes in character form and appending it in the table as shown

Value_Mapping3 = {'Graduate' :1, 'Not Graduate' : 0}
df['Edu_Section'] = df['Education'].map(Value_Mapping3)
df.head(10)
```

```python
df.info()
```

```python
#filling all the empty spaces of the following columns having binary outcomes in tnteger form 
df['Married_Section'].fillna(df.Married_Section.mean(), inplace = True)
df['Gender_Section'].fillna(df.Gender_Section.mean(), inplace = True)
df['Loan_Amount_Term'].fillna(df.Loan_Amount_Term.mean(), inplace = True)
df['Credit_History'].fillna(df.Credit_History.mean(), inplace = True)
df.info()
```

```python
Value_Mapping4 = {'Yes' :1, 'No' : 0}
df['Employed_Section'] = df['Self_Employed'].map(Value_Mapping4)
df.head(10)
```

```python
df.info()
```

```python
df['Employed_Section'].fillna(df.Employed_Section.mean(), inplace = True)
df.info()
```

```python
#Filling the empty spaces of the column having more than 2 outcome in charcter form

lb_make = LabelEncoder()
df['Property_Section'] = lb_make.fit_transform(df['Property_Area'])
df.head(10)
```

```python
Value_Mapping5 = {'Y': 1, 'N' :0}
df['Loan_Section'] = df['Loan_Status'].map(Value_Mapping5)
df.head(10)
```

```python
df.info()
```

```python
##Logistic Reggression##
```

```python
#importing libraries and classes
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#x is the input and y is the output

X = df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History', 'Married_Section', 'Gender_Section', 'Edu_Section', 'Employed_Section', 'Property_Section']].values
y = df[['Loan_Section']].values
```

```python
# importing libraties and classes
#dividing the data in 7:3 ratio for training and testing respectively 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
#Training the model

model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
#Testing Accuracy 

model.score(X_test, y_test)
```

```python
#Y contains all the outputs and x contains all the inputes. i will test on the machine if it gives to expected oupput fotr the
# corresponding inputs.
 

expected = y_test
predicted = model.predict(X_test)
```

```python
#Imoprting Libraries and class
from sklearn import metrics
```

```python
#printing the report
print(metrics.classification_report(expected, predicted))
```

```python
#out of 51 "Y" outcome, 23 were right and 28 were wrong similarly for N 131 right and 3 were wrong

print(metrics.confusion_matrix(expected, predicted))
```
